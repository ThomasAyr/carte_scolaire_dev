Chargement fichier et bibliothèques

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier CSV dans un DataFrame
data_carte_scolaire = pd.read_csv("../../fr-en-occitanie-carte-scolaire-des-colleges-lycees-publics (1).csv", sep=";")

#Source de données : https://data.occitanie.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-occitanie-carte-scolaire-des-colleges-lycees-publics/exports/csv?lang=fr&timezone=Europe%2FBerlin&use_labels=true&delimiter=%3B
```

```python
data_carte_scolaire.head()
```

Statistiques descriptives

```python
# Dimensions du DataFrame
print(f"Nombre de lignes : {data_carte_scolaire.shape[0]}")
print(f"Nombre de colonnes : {data_carte_scolaire.shape[1]}")

# Statistiques descriptives
data_carte_scolaire.describe(include='all')
```

```python
# Compter les valeurs manquantes par colonne
missing_values = data_carte_scolaire.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# Identifier les colonnes catégorielles
categorical_columns = data_carte_scolaire.select_dtypes(include='object').columns

# Afficher les valeurs uniques pour chaque colonne catégorielle
for col in categorical_columns:
    print(f"{col} : {data_carte_scolaire[col].nunique()} valeurs uniques")
    print(data_carte_scolaire[col].value_counts().head())
```

Visualisation des points géographiques (= établisements scolaires)

```python
# Séparer les coordonnées en latitude et longitude
data_carte_scolaire[['latitude', 'longitude']] = data_carte_scolaire['position'].str.split(',', expand=True)

# Convertir les valeurs en type numérique
data_carte_scolaire['latitude'] = pd.to_numeric(data_carte_scolaire['latitude'], errors='coerce')
data_carte_scolaire['longitude'] = pd.to_numeric(data_carte_scolaire['longitude'], errors='coerce')

# Vérifier les premières lignes des colonnes latitude et longitude
print(data_carte_scolaire[['latitude', 'longitude']].head())


plt.figure(figsize=(10, 6))
plt.scatter(data_carte_scolaire['longitude'], data_carte_scolaire['latitude'], alpha=0.5)
plt.title("Répartition géographique des établissements")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
data_carte_scolaire_export = data_carte_scolaire

# Exporter sans numero_de_voie, parite, libelle_commune, position, adresse_uai, appellation_officielle, geo_shape, date_ouverture, ips_etablissement,lieu_dit, latitude,	longitude
data_carte_scolaire_export = data_carte_scolaire_export.drop(columns=['numero_de_voie', 'libelle_commune', 'position', 'adresse_uai', 'appellation_officielle', 'geo_shape', 'date_ouverture', 'ips_etablissement','lieu_dit', 'latitude', 'longitude', 'indice_de_repetition_debut', 'indice_de_repetition_fin'])

# Transformer les colonnes no_de_voie_debut et no_de_voie_fin en entiers, enlever la virgule à code_region et code_academie
data_carte_scolaire_export['no_de_voie_debut'] = pd.to_numeric(data_carte_scolaire_export['no_de_voie_debut'], errors='coerce').fillna(0).astype(int)
data_carte_scolaire_export['no_de_voie_fin'] = pd.to_numeric(data_carte_scolaire_export['no_de_voie_fin'], errors='coerce').fillna(0).astype(int)
data_carte_scolaire_export['code_region'] = pd.to_numeric(data_carte_scolaire_export['code_region'], errors='coerce').fillna(0).astype(int)
data_carte_scolaire_export['code_academie'] = pd.to_numeric(data_carte_scolaire_export['code_academie'], errors='coerce').fillna(0).astype(int)


data_carte_scolaire_export.head()
```

```python
data_carte_scolaire_export.to_csv('../datasets/data_carte_scolaire_nettoye.csv', index=False)
```
